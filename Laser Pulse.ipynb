```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.title('Laser Pulse in Time')

t = np.linspace(-5, 5 , 1000)
I0 = 1
w1 = 2
w2 = w1 *2
I1 = I0 * np.exp((-2 * t**2) / w1 **2)
I2 = I0 * np.exp((-2 * t**2) / w2 **2)

plt.xlabel('Time (ps)')
plt.ylabel('Intensity')

plt.plot(t,I1, label = 'w1 = 2')
plt.plot(t,I2, label = 'w2 = 4')

plt.legend()

plt.show()
```
