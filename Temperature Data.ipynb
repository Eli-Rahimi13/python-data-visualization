```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
temp = np.random.randint(-10, 40, 30)
days = np.arange(1, 31)    
plt.title('Temperature Data')

plt.ylabel('Temperature')
plt.xlabel('Days')

plt.xticks(days[::5])
plt.yticks(np.arange(-10, 41, 5))   

average = np.mean(temp)   

plt.plot(days, temp, label = 'Temp')
plt.axhline(y = average, label = 'average', color = 'red')

plt.legend()

plt.show()
```
