```python
import numpy as np
import matplotlib.pyplot as plt


w0 = 0.001            # beam waist at z=0 (1mm)
wavelenght = 1064e-9     # Nd:YAG
zR = np.pi * w0**2 / wavelenght   # Rayleigh range

z = np.linspace(-3, 3, 1000)

w = w0 * (1 + (z/zR)**2) ** 0.5

# beam profile at z=0

r = np.linspace(-0.003, 0.003, 1000)     # in order to showing the distance well we put r = 3 * w0!
I = np.exp(-2 * r**2 /w0**2)                 ## because in r = w0 intensity would be e2/1


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# subplot1 - beam waist vs distance

ax1.plot(z, w)
ax1.set_title('Beam Waist vs Distance')
ax1.set_xlabel('Distance (m)')
ax2.set_ylabel('Beam Waist (m)')

# subplot2 - beam profile at z=0

ax2.plot(r, I)
ax2.set_title('Beam Profile at z=0')
ax2.set_xlabel('r (m)')
ax2.set_ylabel('Intensity')

plt.tight_layout()

plt.show()
```
