```python
arr = np.linspace(0, 2*np.pi, 1000)

sine = np.sin(arr)
cosine = np.cos(arr)

plt.plot(cosine, color = 'r', label = 'Cos')
plt.plot(sine, color = 'b', label = 'Sin')

plt.grid()
plt.legend()

plt.show()
```
