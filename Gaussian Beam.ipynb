```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
plt.title('Gaussian Beam')

r = np.linspace(-4, 4, 1000)
I0 = 1    
w = 2    

I = I0 * np.exp((-2 * r **2) / w **2)

plt.plot(r,I)

plt.xlabel('Distance from the Beam')
plt.ylabel('Intensity')

plt.show()

```
